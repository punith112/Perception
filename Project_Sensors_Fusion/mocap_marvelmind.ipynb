```python
def compare_mocap(marvel_pose, vicon_pose):
    plt.plot( vicon_pose.x-vicon_pose.x[0], vicon_pose.y-vicon_pose.y[0] )
    plt.plot( marvel_pose.x_m-marvel_pose.x_m[0], - (marvel_pose.y_m-marvel_pose.y_m[0]) )
    plt.legend(['mocap', 'ultrasound'])
```

```python
from mpl_toolkits.mplot3d import Axes3D

def plot3D(marvel_pose3D, vicon_pose3D):
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(111, projection='3d')
    plt.xlabel('X'); plt.ylabel('Y');

    ax.plot( (marvel_pose3D.y_m-marvel_pose3D.y_m[0]), -( marvel_pose3D.x_m-marvel_pose3D.x_m[0] ), -(marvel_pose3D.z_m-marvel_pose3D.z_m[0]))
    ax.plot(vicon_pose3D.x-vicon_pose3D.x[0], vicon_pose3D.y-vicon_pose3D.y[0], vicon_pose3D.z-vicon_pose3D.z[0])
```

```python
plot3D(marvel_pose3D, vicon_pose3D)
```
